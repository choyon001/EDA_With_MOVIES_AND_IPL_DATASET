```python
def track_record(team1,team2):
    team1_mask = (ipl['Team1']==f'{team1}' ) & (ipl['Team2']==f'{team2}') 
    team2_mask = (ipl['Team1']==f'{team2}' ) & (ipl['Team2']==f'{team1}')
    matches = ipl[team1_mask | team2_mask]
    total = matches['WinningTeam'].shape[0]
    team1_wins = matches[matches['WinningTeam'] == f'{team1}'].shape[0]
    
    print(f"{team1} wins {team1_wins} matches out of {total} matches against {team2}")
    
```

```python
track_record('Royal Challengers Bangalore','Kolkata Knight Riders')
```
